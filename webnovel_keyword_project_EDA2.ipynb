#### '로맨스판타지'는 모든 작품의 키워드로 들어가있음. 제거해줌.

```python
def remove_keyword(keyword_list, keyword_to_remove):
    li = []
    for keyword in keyword_list:
        if keyword != keyword_to_remove:
            li.append(keyword)
    return li

df['r키워드'] = ''
for index, row in df.iterrows():
    keyword_list = row['키워드']
    removed_keywords = remove_keyword(keyword_list, '로맨스판타지')
    df.at[index, 'r키워드'] = removed_keywords

```

#### 두 개의 키워드를 선택하여 Counter() 객체 이용하여 공통 등장 횟수 계산

```python
from itertools import combinations
from collections import Counter
keyword_lists = df['r키워드'].tolist()

keyword_counts = Counter()

for keyword_list in keyword_lists:
    for pair in combinations(keyword_list, 2):
        keyword_counts[pair] += 1

print(keyword_counts)
```

```python
filtered_items = {key: value for key, value in keyword_counts.items() if value > 13}


sorted_items = sorted(filtered_items.items(), key=lambda x: x[1], reverse=True)
```

```python
sorted_items
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from collections import Counter
# font_manager._rebuild()
plt.rc('font', family='NanumGothic')

keywords = list(set(keyword for keywords, _ in sorted_items for keyword in keywords))


matrix = [[0 for _ in keywords] for _ in keywords]

for (keyword1, keyword2), count in sorted_items:
    index1 = keywords.index(keyword1)
    index2 = keywords.index(keyword2)
    matrix[index1][index2] = count
    matrix[index2][index1] = count

plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=True, fmt='d', cmap='viridis', xticklabels=keywords, yticklabels=keywords)
plt.xlabel('키워드')
plt.ylabel('키워드')
plt.title('키워드 간의 빈도수 히트맵')
plt.show()
```

### 연관규칙 알고리즘을 위해 one-hot encoding

```python
all_keywords = set(keyword for sublist in df['키워드'] for keyword in sublist)
binary_data = {}
for keyword in all_keywords:
    binary_data[keyword] = df['키워드'].apply(lambda x: 1 if keyword in x else 0)

df_binary = pd.DataFrame(binary_data)

print(df_binary)
```

```python

correlation_matrix = df_binary.corr()

print(correlation_matrix)
```

```python

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap="coolwarm", vmin=-1, vmax=1, annot=False, fmt=".2f", linewidths=0.5)
plt.title('상관 행렬 히트맵')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

df_binary = pd.read_csv('binary_data.csv')

frequent_itemsets = apriori(df_binary, min_support=0.1, use_colnames=True)


rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print(rules)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="support", y="confidence", size="lift", data=rules)
plt.title("Association Rules")
plt.show()
```
